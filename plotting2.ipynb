```python
import matplotlib.pyplot as plt
import csv

files = ["3q-1000samples.csv","4q-1000samples.csv","5q-1000samples.csv"]

data_expr = {}
data_ent = {}

for filename in files:
    n = int(filename[0])
    data_expr[n] = {}
    data_ent[n] = {}
    for L in range(1,5):
        data_expr[n][L]={}
        data_ent[n][L]={}
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)

        # Use a counter to track the current line number
        line_count = 0

        # Loop through the rows in the CSV file
        for row in csv_reader:
            line_count += 1
            if line_count <= 3:
                continue

            id = row[0].partition("-")[0]
            L = int(row[1])
            expr = float(row[2])
            ent = float(row[3])
            t = float(row[4])
            print(n, L, id, sep=" ")

            #data_expr[n][L] = data_expr.get(L, {})
            data_expr[n][L][id] = expr

            #data_ent[n][L] = data_ent.get(L, {})
            data_ent[n][L][id] = ent
```

```python
# Plotting expressibility

L = 4

my_xticks = list(data_expr[5][1].keys())
x = [i+1 for i in range(len(my_xticks))]

def comp_expr(circuit):
    return data_expr[3][L][circuit]

def comp_ent(circuit):
    return data_ent[3][1][circuit]


my_xticks.sort(key=comp_expr)

markers=["o", "s", "*", "d", "p", "v", "h", "^"]
colors = ["r", "b", "k", "g", "m", "c", "y"]


for n in range(3, 6):
    y = [data_expr[n][L][c] for c in my_xticks]
    plt.scatter(x,y,label=f"n = {n}", marker=markers[n-1], facecolors="none", edgecolors=colors[n-1])

plt.xlabel('Circuit ID')
plt.ylabel('Expr')
plt.yscale('log')
plt.xticks([i+1 for i in range(len(my_xticks))], my_xticks)
plt.legend()
plt.title(f"{L} layers circuits") 
plt.show()
```
